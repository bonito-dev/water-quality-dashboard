```python
from dotenv import load_dotenv
import psycopg2, os

load_dotenv("../.env")

# Try pooler connection - port 6543
conn = psycopg2.connect(
    host=os.getenv("DB_HOST"),
    port=6543,                    # pooler port
    dbname=os.getenv("DB_NAME"),
    user=os.getenv("DB_USER"),
    password=os.getenv("DB_PASSWORD"),
    sslmode="require"
)
print("Connected:", conn.status)
conn.close()
```

```python
import pandas as pd
from pathlib import Path
import datetime

# Go up one level from notebooks/ to the project root
PROCESSED = Path("../data/processed")

df = pd.read_csv(PROCESSED / "jmp_kenya_clean.csv")
print("Columns:", df.columns.tolist())
print("Shape:", df.shape)
```

```python
CURRENT_YEAR = datetime.date.today().year

df["is_estimated"] = df["year"] > CURRENT_YEAR

print("is_estimated value counts:")
print(df["is_estimated"].value_counts())

df.to_csv(PROCESSED / "jmp_kenya_clean.csv", index=False)
print("Resaved with is_estimated column")
```
